```python
import subprocess
process = subprocess.Popen(
    [
        "ssh",
        "-v",   # or -vv / -vvv for more verbosity
        "-T",
        "-p", "443",
        "-R0:localhost:11434",
        "a.pinggy.io",
        "-o", "StrictHostKeyChecking=no",
        "-o", "UserKnownHostsFile=/dev/null"
    ],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)
```

```python
for line in process.stdout:
    print(line, end="")
```
